```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv('machine.data',
                         index_col=None,
                         header=None,
                         names=["vendor name","model name","MYCT","MMIN","MMAX","CACH","CHMIN","CHMAX","PRP","ERP"])
```

```python
data_frame
```

```python
def get_model_myct(df):
    df = data_frame
    df_mean = pd.pivot_table(data=df,values=df.columns[2:],aggfunc=np.mean,columns='vendor name')
    df_median = pd.pivot_table(data=df,values=df.columns[2:],aggfunc=np.median,columns='vendor name')

    df_mean_myct = pd.DataFrame({'vendor name':list(df_mean.columns),"Mean":df_mean.values.tolist()[6]})
    df_median_myct = pd.DataFrame({'vendor name':list(df_median.columns),"Median":df_median.values.tolist()[6]})

    df_melt_model = pd.melt(frame=df,value_vars="model name",id_vars=["vendor name"])
    df_melt_mean = pd.melt(frame=df_mean_myct,value_vars='Mean',id_vars=['vendor name'])
    df_melt_median = pd.melt(frame=df_median_myct,value_vars='Median',id_vars=['vendor name'])

    final_df = pd.concat([df_melt_model,df_melt_mean,df_melt_median],ignore_index=True)
    return final_df
```

```python
data_model = get_model_myct(data_frame)
```
