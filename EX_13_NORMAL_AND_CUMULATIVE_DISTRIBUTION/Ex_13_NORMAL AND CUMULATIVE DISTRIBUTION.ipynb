# Normal Distribution

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def fn_norm_check(test_data):
    height = np.linspace(150,180,100)
    plt.hist(height,12)
    plt.show()

    mean_height = np.mean(height)
    std_height = np.std(height)

    pdf_height = norm.pdf(height,mean_height,std_height)

    plt.plot(pdf_height)
    plt.xlabel("Adult Height")
    plt.ylabel("Probablility distribution")
    plt.show()

    pdf_test = norm.pdf(test_data,mean_height,std_height)
    print(pdf_test)
    if pdf_test >=min(pdf_height) and pdf_test <= max(pdf_height):
        print("test data is adult height")
    else:
        print("test data is not an adult height")
```

```python
fn_norm_check(170)
```

# Cumulative Distribution

```python
from scipy.stats import norm
import numpy as np
```

```python
def fn_cumulative_dist(test_data1,test_data2):
    height = np.linspace(150,180,100)
    
    mean_h = np.mean(height)
    std_h = np.std(height)
    
    t1_cdf = norm(loc=mean_h,scale=std_h).cdf(test_data1)
    t2_cdf = norm(loc=mean_h,scale=std_h).cdf(test_data2)
    
    
    print(f"Probability of height to be under {test_data1} is {t1_cdf}")
    print(f"Probabitly of height between {test_data1} and {test_data2} is {t2_cdf-t1_cdf}")
    print(f"Probability of height above {test_data2} is {1-t2_cdf}")
```

```python
test_data1 = 160
test_data2 =170
fn_cumulative_dist(test_data1,test_data2)
```
