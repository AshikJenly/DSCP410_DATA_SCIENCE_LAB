```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../Dataset/machine.data',
                 index_col=None,
                 header=None,
                 names=["vendor name","Model Name","MYCT","MMIN","MMAX","CACH","CHMIN","CHMAX","PRP","ERP"])
```

```python
df
```

```python
df_mean_sub = pd.pivot_table(data=df,values=["MMIN","MMAX"],columns="vendor name",aggfunc=np.mean)
```

```python
df_mean_sub
```

```python
sns.heatmap(data=df_mean_sub)
```
