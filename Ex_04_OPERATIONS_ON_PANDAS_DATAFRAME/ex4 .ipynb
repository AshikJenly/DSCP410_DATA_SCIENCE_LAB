```python
import numpy as np
import pandas as pd
```

```python

def get_r_name():
    first_name = ['ashik','jenly','salman','khan']
    last_name = ['kaif','swen','saif','ligith']
    
    return f"{np.random.choice(first_name)} {np.random.choice(last_name)}"
```

```python
data = {
    'Roll':np.arange(101,111),
    'Name':[get_r_name() for _ in range(0,10)],
    'Attend8':np.random.randint(50,100,10),
    'Attend9':np.random.randint(0,101,10),
    'Attend10':np.random.randint(5,101,10),
    'Assignment':np.random.choice([5,14,15],10),
    'Test':np.random.randint(0,20,10)
}
```

```python
df=pd.DataFrame(data)
display(df)
display(df.isnull().sum(axis=0))
```

```python
df_gt_80 = df[df['Attend10']>80]
df_bt_70_80 = df[(df['Attend10']>70) & (df['Attend10'] <80)]
df_lt_80 = df[df['Attend10']<=70]
```

```python
display(df.sort_values(by='Attend9',ascending=False))
```

```python
df['consolidated_attend']=(df['Attend10'] + df['Attend9'] + df['Attend8'])/3
display(df[df['consolidated_attend']>60])
```

```python
df['Result']=df['Test'] + df['Assignment']
df['Result']=df['Result'].apply(lambda x : 'Pass' if x>25 else 'Fail')
```
