```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = sm.datasets.copper.load_pandas().data
```

# LINE CHART

### Plot line chart to display copper and aluminium prices during 1950 - 1975

```python
fig,ax = plt.subplots(2,1)

ax[0].set(title='Copper and Aluminium',ylabel='Copper Price')
ax[0].plot(df['COPPERPRICE'].index + 1950,df['COPPERPRICE'],'r--')


ax[1].set(xlabel='Time',ylabel='Aluminium Price')
ax[1].plot(df['ALUMPRICE'].index + 1950 ,df['ALUMPRICE'],'b--')
```

# Bar Chat

### Plot using bar and line chart in same figure to depict the world consumption and manufacturing inventory

```python
fig,ax1 = plt.subplots(figsize=(15,8))

ax2 = ax1.twinx()

x = df['WORLDCONSUMPTION'].index + 1950

ax1.bar(x,df['WORLDCONSUMPTION'],color='cyan')
ax1.set(ylabel='Word Consumption',xlabel="Time")

ax2.plot(x,df['INVENTORYINDEX'],'r-*',label='Manufacturing Inventory')

ax2.legend(loc='upper left')

plt.show()
```
